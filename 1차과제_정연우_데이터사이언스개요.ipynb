## **1. Chat-GPT로 코딩하기 (20점)**



다음은 미로의 입구에서 출발해서 유저 인풋 기반으로 보물을 찾는 보물 찾기 게임의 소스 코드이다.

직접 게임을 플레이하며 여기에서 발생하는 오류 사항을 찾고, **Chat-GPT만 사용해서** 고치시오. (Chat-GPT가 작성한 코드 복사 붙여넣기만 허용)
